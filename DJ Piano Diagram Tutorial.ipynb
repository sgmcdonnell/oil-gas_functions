```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

from matplotlib.patches import Ellipse
```

**Setting up plot:**

```python
    #DP 129
pad = '129_'

fig, ax = plt.subplots()
plt.gcf().set_size_inches(14, 5)

xmax = 90
ymax = 9
xr = 7

ax.set(xlim=(xmax, 0), ylim = (0, ymax))
```

**Formatting labels/tick marks on plot:**

```python
ax.set(xlim=(xmax, 0), ylim = (0, ymax))

    #plt.yticks(np.arange(start,stop, step),[labels])
plt.yticks(np.arange(0.5, ymax, step=1),['Codell', 'Fort Hays','Nio D', 'C Marl', 'C Chalk','B Marl', 'B Chalk', 'A Marl','A Chalk'])
ax.tick_params(axis='y', labelsize=18)

    #plt.yticks(np.arange(start,stop, step),[labels])
plt.xticks(np.arange(90, -1, step=-5),['90', '85','80','75', '70', '65','60', '55', '50','45','40','35','30','25','20','15','10'])
ax.tick_params(axis='x', labelsize=18)
```

**Adding section lines:**

```python
    #plotting vertical black lines to show section lines
    #plt.vlines(x,ymin, ymax, line parameters)
plt.vlines(90, 0, ymax, colors='black', linestyles='solid', label='',linewidth=6)
plt.vlines(10, 0, ymax, colors='black', linestyles='solid', label='',linewidth=6)
```

**Adding colored rectangles for each formation:**

```python
fig, ax = plt.subplots()
plt.gcf().set_size_inches(14, 5)
ax.set(xlim=(90, 0), ylim = (0, ymax))
plt.yticks(np.arange(0.5, ymax, step=1),['Codell', 'Fort Hays','Nio D', 'C Marl', 'C Chalk','B Marl', 'B Chalk', 'A Marl','A Chalk'])
ax.tick_params(axis='y', labelsize=18)


#adding in color rectangles 
    #ax.add_patch(patches.Rectangle((xmin, ymin), w_rect, h_rect))
ax.add_patch(patches.Rectangle((0, 0), xmax+xr, 1, color='wheat'))
ax.add_patch(patches.Rectangle((0, 1), xmax+xr, 1, color = 'powderblue'))
ax.add_patch(patches.Rectangle((0, 2), xmax+xr, 1, color = 'lightgray'))
ax.add_patch(patches.Rectangle((0, 3), xmax+xr, 2, color = 'lemonchiffon'))
ax.add_patch(patches.Rectangle((0, 5), xmax+xr, 2, color = 'lightskyblue'))
ax.add_patch(patches.Rectangle((0, 7), xmax+xr, 2, color = 'pink'))
```

**Adding gridlines**

```python
fig, ax = plt.subplots()
plt.gcf().set_size_inches(14, 5)
ax.set(xlim=(90, 0), ylim = (0, ymax))
plt.yticks(np.arange(0.5, ymax, step=1),['Codell', 'Fort Hays','Nio D', 'C Marl', 'C Chalk','B Marl', 'B Chalk', 'A Marl','A Chalk'])
ax.tick_params(axis='y', labelsize=18)

#formatting gridlines
ax.xaxis.grid(zorder=0,color ='black', linestyle = ':')
ax.grid(which='minor', linestyle=':', linewidth='1', color='black')
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
```
